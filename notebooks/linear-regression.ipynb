```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def train_linear_regression_sgd(X,y,lr,epochs):
    X = np.asarray(X,dtype=float)
    y = np.asarray(y,dtype=float)
    
    w = 0.0
    b = 0.0
    
    for epoch in range(epochs):
        for i in range(len(X)):
            prediction = w*(X[i])
            y_hat = prediction + b
            error = prediction - y[i]
            w = w - (lr*error*X[i])
            b = b - (lr*error)
            
    return w,b  
```

```python
def train_linear_regression_bgd(X,y,lr = 0.01, epochs = 1000):
    X = np.asarray(X,dtype=float)
    y = np.asarray(y,dtype=float)
    
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0.0
    losses = []
    
    for epoch in range(epochs):
        # predictions
        y_hat = X @ w + b 
        
        # Errors
        error = y_hat - y
        
        # Gradients
        grad_w = (X.T @ error)/n_samples
        grad_b = error.mean()
        
        # weight updates
        w = w - lr*grad_w
        b = b - lr*grad_b
        
        # losses
        loss = np.mean(error**2)
        losses.append(loss)
    return w, b, losses
```

```python
X = np.array([1, 2, 3, 4], dtype=float)
y = np.array([5, 7, 9, 11], dtype=float)  # y = 2x + 3

w, b = train_linear_regression_sgd(X, y, lr=0.01, epochs=2000)
print("SGD -> w:", w, "b:", b)
```
